```python
import pandas as pd
import plotly.express as px

df = pd.read_csv('https://raw.githubusercontent.com/qed747/testdata/main/OFAC_Plotly.csv?raw=True', encoding='ISO-8859-1')


fig = px.scatter(df,
                 x='Announcement Date',
                 y='Penalty',
                 hover_name='Industry',
                 title='Cereal ratings vs. sugars')

fig.show()
```

```python
# Remove dollar sign and commas, then convert to float
df['Penalty'] = df['Penalty'].replace({'\$': '', ',': ''}, regex=True).astype(float)

# Now group by 'Nationality' and 'Industry' and sum 'Penalty'
df_grouped = df.groupby(['Nationality','Industry'])['Penalty'].sum().reset_index()

# Plot the sunburst
fig = px.sunburst(df_grouped, path=['Nationality','Industry'], values='Penalty')
fig.show()
```
